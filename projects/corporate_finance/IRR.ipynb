### Net present value og Internal rate of return

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Net present value function:
def npv(cf, r):
    NPV = cf[0]
    for i in range(1, len(cf)):
        NPV += cf[i]/((1 + r)**i)
    return(NPV)

#Parameters:
###############################################
# Cash flow
cf1 = (-3000, 1815, 1815, 1815, 1815, 1815)
cf2 = (-2000, 1000, 1000, 1000, 1000, 1000)
cf_list = [cf1, cf2]

# Investment names:
names = ['Project A', 'Project B']
# Risk free discount rate:
rate_list = [0.10, 0.15]
###############################################

# Calculating Net present value:
for rate in rate_list:
    for i, cf in enumerate(cf_list):
        print(f'{names[i]} with {rate*100}% risk free discount rate:')
        print(f'Net present value: {npv(cf, rate):.0f}\n')

# Calculating and plotting internal rate of return:
fig, axs = plt.subplots(len(cf_list), 1, figsize=(10,4 * len(cf_list)))

for i, cf in enumerate(cf_list):
    r_list = np.linspace(0, 1, 1000)
    NPV_list = list()
    for r in r_list:
        NPV_list.append(npv(cf, r))
    poly = np.polyfit(np.array(NPV_list), r_list, deg=20)
    irr = np.polyval(poly, 0)
    print(f'{names[i]} Internal rate of return: {irr * 100:.1f}%\n')
    
    # NPV curve plot with IRR point:
    axs[i].plot(r_list, NPV_list, 'r')
    axs[i].scatter(irr, 0)
    axs[i].xaxis.set_major_formatter(mtick.PercentFormatter(1))
    axs[i].set_ylabel('NPV', fontsize=14)
    axs[i].set_xlabel('Rate', fontsize=14)
    axs[i].set_title(f'{names[i]} Internal rate of return point', fontsize=15)
    axs[i].grid()
    axs[i].axhline(y=0, linestyle='--', linewidth=1.2)
    fig.tight_layout();
```
