## Notebook for CSV data creation

```python

# Salmon data from SSB
salmon_data_URL = 'https://www.ssb.no/statbank/sq/10086603'
salmon_data = import_salmon_data(salmon_data_URL, rename_cols=True)

# Commodity data from yahoo finance
commodity_data = yf.download(
    ['LE=F', 'HE=F', 'ZW=F', 'ZS=F'],
    start=salmon_data.index[0], 
    end=salmon_data.index[-1], 
    interval='1wk', 
    progress=False
)['Adj Close']

data = pd.concat(
    [
        salmon_data, 
        commodity_data.reindex(salmon_data.index)
    ],
    axis=1
).dropna().rename(
    columns={
        'LE=F': 'Cattle_price', 
        'HE=F': 'Pork_price', 
        'ZW=F': 'Wheat_price', 
        'ZS=F': 'Soybean_price', 
        'Price': 'Salmon_price',
        'Volume': 'Salmon_volume'
    }
)

display(data)

# Save data
data.to_csv('data_files/complete_data.csv')
```
