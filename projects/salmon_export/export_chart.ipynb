```python
import requests
import pandas as pd
```

```python

export_URL = 'https://oec.world/olap-proxy/data?cube=trade_i_baci_a_92&drilldowns=Year,Exporter Country&measures=Trade Value&parents=true&Year=2021&HS6=1030212&properties=Exporter Country ISO 3'
import_URL = 'https://oec.world/olap-proxy/data?cube=trade_i_baci_a_92&drilldowns=Year,Importer Country&measures=Trade Value&parents=true&Year=2021&HS6=1030212&properties=Importer Country ISO 3'

req_import = requests.get(export_URL).json()
req_export = requests.get(import_URL).json()

country_list = list()
export_value_list = list()
import_value_list = list()

for export_data, import_data in zip(req_import['data'], req_export['data']):
    country_list.append(export_data['Country'])
    export_value_list.append(export_data['Trade Value'])
    import_value_list.append(import_data['Trade Value'])


df = pd.DataFrame(
    {
        'Country': country_list, 
        'Export value': export_value_list, 
        'Import value': import_value_list
    }
).sort_values('Export value', ascending=False).reset_index(drop=True)

display(df)
```
